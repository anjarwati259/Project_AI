```python
#correlation matric untuk numeric value (umur, berar akhir, jmlh tahun pendidikan, Keuntungan Kapital, 
#Kerugian Capital, Jam per Minggu)
g = sns.heatmap(df[['Gaji', 'Umur', 'Berat Akhir', 'Jmlh Tahun Pendidikan','Keuntungan Kapital','Kerugian Capital','Jam per Minggu']].corr(),
               annot=True, fmt = '.2f', cmap = 'coolwarm')
```

```python
#gaji dengan umur
g = sns.catplot(x='Jmlh Tahun Pendidikan', y="Pendidikan", data=df,kind='bar', height=6, palette='vlag')
g.despine(left=True)
g = g.set_ylabels('Probabilitas Gaji')
```
