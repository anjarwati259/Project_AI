# Tentang Data Set

Terdapat tiga file yang digunakan pada final project ini, file train.csv berisi dataset yang memiliki kolom label digunakan untuk traning model, sedangkan file test.csv berisi dataset tanpa kolom label, yang nantinya teman-teman dapat menambahkan kolom 'Gaji' berisi prediksi teman-teman sesuai ketentuan pada tab Overview. Kemudian teman-teman diminta untuk membuat dataset baru berisi kolom 'id' dan kolom 'Gaji' yang diambil dari dataset test yang telah teman-teman tambahkan prediksinya, dataset baru ini yang nantinya akan dijadikan file submission. Contoh isi dari submission dapat teman-teman lihat pada file sampleSubmission.csv

- id - id unique yang dimiliki setiap sample
- Umur - umur yang dimiliki setiap sample
- Kelas Pekerja - kelompok kelas pekerjaan masing-masing sample
- Berat Akhir - berisi nilai akumulasi berdasarkan populasi, ras, dan gender dengan umur 16+ suatu wilayah, sample yang diambil dari wilayah yang memiliki karakteristik demografis yang sama akan memiliki nilai berat akhir yang sama
- Pendidikan - tingkat pendidikan terakhir masing-masing sample
- Jmlh Tahun Pendidikan - berisi jumlah tahun masing-masing sample mengenyam pendidikan
- Status Perkawinan - status perkawinan masing-masing sample
- Pekerjaan - pekerjaan saat ini masing-masing sample
- Jenis Kelamin - jenis kelamin masing-masing sample
- Keuntungan Kapital - keuntungan yang didapat jika sample menjual semua aset miliknya
- Kerugian Kapital - kerugian yang didapat jika sample menjual semua aset miliknya
- Jam per Minggu - jam kerja masing-masing sample setiap minggunya
- Gaji - nilai gaji masing-masing sample apakah kurang dari sama dengan 5jt atau lebih dari 5jt

```python
import pandas as pd
import numpy as np
```

```python
#load data
df_train = pd.read_csv('train.csv')
df_test = pd.read_csv('test.csv')
idx = len(df_train)
```

```python
# Checking train and test sets
display(df_train.sample(3))
display(df_test.sample(3))
```

```python
# Merging data
df_train.drop('id', axis=1, inplace=True)
df_test.drop('id', axis=1, inplace=True)
df_merged = pd.concat([df_train, df_test], sort=False).reset_index(drop=True)
df_merged
```

```python
display(df_merged.columns)
```

```python
df_merged.replace('?', np.nan, inplace=True)
```

```python
#cek missing value
total = df_merged.isna().sum()
persen = round(total*100/len(df_train),2)
missing_value=pd.DataFrame({'Total':total, '%':persen})
missing_value.sort_values('%', inplace=True, ascending=False)
missing_value.head()
```

```python
#replace missing value dengan Spesialis
df_merged['Pekerjaan'] = df_merged['Pekerjaan'].fillna(method='bfill')
```

```python
#replace missing value dengan Wiraswasta
df_merged['Kelas Pekerja'] = df_merged['Kelas Pekerja'].fillna(method='ffill')
```

```python
#cek missing value lagi
df_merged.isna().sum()
```

```python
#ubah type data menjadi category
df_merged['Kelas Pekerja'] = df_merged['Kelas Pekerja'].astype('category')
df_merged['Pekerjaan'] = df_merged['Pekerjaan'].astype('category')
#df_merged['Status Perkawinan'] = df_merged['Status Perkawinan'].astype('category')
df_merged.dtypes
```

```python
#ubah data dengan .cat.codes
df_merged['Kelas Pekerja'] = df_merged['Kelas Pekerja'].cat.codes
df_merged['Pekerjaan'] = df_merged['Pekerjaan'].cat.codes
#df_merged['Status Perkawinan'] = df_merged['Status Perkawinan'].cat.codes
df_merged.head()
```

```python
df_merged = pd.get_dummies(df_merged, columns=['Status Perkawinan'], prefix = ['Status'])
```

```python
#lihat coloms pendidikan
df_merged['Pendidikan'].value_counts()
```

```python
#ubah data pedidikan
param = ['1st-4th','5th-6th','7th-8th','9th','10th','11th','12th']
for i in param:
    df_merged['Pendidikan'].replace(i, 'Tidak Lulus Kuliah', inplace=True)
```

```python
# encoding ordinal data
param = {'SD':1,'SMP':1,'SMA':2,'Tidak Lulus Kuliah':3,
         'Sekolah Professional':4,'Pendidikan Tinggi':5,
         'D3':6,'D4':7,'Sarjana':8,'12th':9,'Master':10,'Doktor':11}
df_merged['Pendidikan'].replace(param, inplace=True)
df_merged.head()
```

```python
#range data umur
df_merged['umur']=0
df_merged.loc[df_merged['Umur']<=18,'umur']=0
df_merged.loc[(df_merged['Umur']>18)&(df_merged['Umur']<=36),'umur']=1
df_merged.loc[(df_merged['Umur']>36)&(df_merged['Umur']<=54),'umur']=2
df_merged.loc[(df_merged['Umur']>54)&(df_merged['Umur']<=72),'umur']=3
df_merged.loc[df_merged['Umur']>72,'umur']=4
df_merged.head()
```

```python
#encoding jam kerja
df_merged['Jam_kerja']=0
df_merged.loc[df_merged['Jam per Minggu']<=20,'Jam_kerja']=0
df_merged.loc[(df_merged['Jam per Minggu']>20)&(df_merged['Jam per Minggu']<=40),'Jam_kerja']=1
df_merged.loc[(df_merged['Jam per Minggu']>40)&(df_merged['Jam per Minggu']<=60),'Jam_kerja']=2
df_merged.loc[(df_merged['Jam per Minggu']>60)&(df_merged['Jam per Minggu']<=80),'Jam_kerja']=3
df_merged.loc[df_merged['Jam per Minggu']>80,'Jam_kerja']=4
df_merged.head(2)
```

```python
#encoding keuntungan capital
df_merged['Keuntungan_Kapital']=0
df_merged.loc[df_merged['Keuntungan Kapital']<=279997200,'Keuntungan_Kapital']=0
df_merged.loc[(df_merged['Keuntungan Kapital']>279997200)&(df_merged['Keuntungan Kapital']<=559994400),'Keuntungan_Kapital']=1
df_merged.loc[(df_merged['Keuntungan Kapital']>559994400)&(df_merged['Keuntungan Kapital']<=839991600),'Keuntungan_Kapital']=2
df_merged.loc[(df_merged['Keuntungan Kapital']>839991600)&(df_merged['Keuntungan Kapital']<=1119988800),'Keuntungan_Kapital']=3
df_merged.loc[df_merged['Keuntungan Kapital']>1119988800,'Keuntungan_Kapital']=4
df_merged.head(2)
```

```python
#encoding kerugian capital
df_merged['Kerugian_Capital']=0
df_merged.loc[df_merged['Kerugian Capital']<=12196800,'Kerugian_Capital']=0
df_merged.loc[(df_merged['Kerugian Capital']>12196800)&(df_merged['Kerugian Capital']<=24393600),'Kerugian_Capital']=1
df_merged.loc[(df_merged['Kerugian Capital']>24393600)&(df_merged['Kerugian Capital']<=36590400),'Kerugian_Capital']=2
df_merged.loc[(df_merged['Kerugian Capital']>36590400)&(df_merged['Kerugian Capital']<=48787200),'Kerugian_Capital']=3
df_merged.loc[df_merged['Kerugian Capital']>48787200,'Kerugian_Capital']=4
df_merged.head(2)
```

```python
#umur
df_merged['Jmlh_thn_Pendidikan']=0
df_merged.loc[df_merged['Jmlh Tahun Pendidikan']<=4,'Jmlh_thn_Pendidikan']=0
df_merged.loc[(df_merged['Jmlh Tahun Pendidikan']>4)&(df_merged['Jmlh Tahun Pendidikan']<=8),'Jmlh_thn_Pendidikan']=1
df_merged.loc[(df_merged['Jmlh Tahun Pendidikan']>8)&(df_merged['Jmlh Tahun Pendidikan']<=12),'Jmlh_thn_Pendidikan']=2
df_merged.loc[df_merged['Jmlh Tahun Pendidikan']>12,'Jmlh_thn_Pendidikan']=4
df_merged.head(2)
```

```python
#replace columns jenis kelamin menjadi biner
df_merged['Jenis Kelamin'].replace(['Perempuan','Laki2'],[0,1],inplace=True)
```

```python
#replace label gaji ke biner
df_merged['Gaji'].replace(['<=7jt','>7jt'],[0,1], inplace = True)
```
